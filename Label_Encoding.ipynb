#Label Encoding
Converting the labels into numeric form

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Dataset

```python
# Loading the data from csv file to pandas dataframe
df = pd.read_csv('/content/breast_cancer_data.csv')
```

```python
df.head()
```

```python
# Count the values of different labels in diagnosis column
df['diagnosis'].value_counts()
```

```python
# Load the label encoder function
label_encoder = LabelEncoder()
```

```python
label = label_encoder.fit_transform(df.diagnosis)
```

```python
# Appending the target column in data frame
df['target'] = label
```

```python
df.head()
```

**0 = Benign**

**1 = Malignant**


```python
# Count the values of different labels in target column
df['target'].value_counts()
```

Label Encoding of Iris Dataset

```python
# Loading the data from csv file to pandas dataframe
iris_data = pd.read_csv('/content/iris_data.csv')
```

```python
iris_data.head()
```

```python
iris_data.tail()
```

```python
# Count the values of different labels in Species column
iris_data['Species'].value_counts()
```

```python
# Load the label encoder function
label_encode = LabelEncoder()
```

```python
label_iris = label_encode.fit_transform(iris_data.Species)
```

```python
# Appending the Species_num column in the data frame
iris_data['Species_Num'] = label_iris
```

```python
iris_data.head()
```

**0 = Iris-setosa**

**1 = Iris-versicolor**

**2 = Iris-virginica**

```python
iris_data['Species_Num'].value_counts()
```
